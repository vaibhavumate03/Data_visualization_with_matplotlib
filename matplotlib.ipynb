```python
sales = [17685,24565,76553,76543,74545,24344,57644,65288]
profit = [1259,3654,4852,1599,6412,5552,1633,2800]
```

```python
com = ['Accenture','IBN','amazon','ITC','EY','Cognizant','Infosys','google']
xpos = np.arange(len(com))
plt.bar(com,sales,color ='orange',label = 'sales', width=0.4)
plt.xticks(xpos,com)
plt.show()
```

```python
plt.bar(xpos,sales,color ='orange',label = 'sales',width=0.4)
plt.bar(xpos,profit,color= 'blue',label= 'profit',width=0.4)
plt.legend()
plt.xticks(xpos,com)
plt.show()
```

```python
plt.bar(xpos+0.2,sales,color ='orange',label = 'sales',width=0.4)
plt.bar(xpos-0.2,profit,color= 'blue',label= 'profit',width=0.4)
plt.legend()
plt.xticks(xpos,com)
plt.show()
```
